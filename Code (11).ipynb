```python
# ============================================================
# STEP 3: DATA PREPROCESSING
# MIN-MAX NORMALIZATION + ONE-HOT ENCODING
# ============================================================

import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler, OneHotEncoder


# ============================================================
# 1. CREATE A COPY OF MERGED DATASET
# ============================================================

preprocessed_df = merged_df.copy()

print("=" * 70)
print("STEP 3: DATA PREPROCESSING")
print("=" * 70)

print("\nOriginal Dataset Shape:", preprocessed_df.shape)


# ============================================================
# 2. DEFINE NUMERICAL FEATURES
# ============================================================

numerical_columns = [

    "age",

    "structured_attention_score",

    "session_duration",

    "text_attention_score",

    "engagement_score",

    "text_length"

]


print("\nNumerical Features:")

for column in numerical_columns:

    print("-", column)


# ============================================================
# 3. DISPLAY VALUES BEFORE NORMALIZATION
# ============================================================

print("\n" + "=" * 70)
print("NUMERICAL FEATURES BEFORE NORMALIZATION")
print("=" * 70)

print(

    preprocessed_df[
        numerical_columns
    ].head()

)


# ============================================================
# 4. INITIALIZE MIN-MAX SCALER
# ============================================================

scaler = MinMaxScaler(
    feature_range=(0, 1)
)


# ============================================================
# 5. APPLY MIN-MAX NORMALIZATION
# ============================================================

preprocessed_df[
    numerical_columns
] = scaler.fit_transform(

    preprocessed_df[
        numerical_columns
    ]

)


# ============================================================
# 6. DISPLAY NORMALIZED VALUES
# ============================================================

print("\n" + "=" * 70)
print("NUMERICAL FEATURES AFTER MIN-MAX NORMALIZATION")
print("=" * 70)

print(

    preprocessed_df[
        numerical_columns
    ].head()

)


# ============================================================
# 7. VERIFY NORMALIZATION RANGE
# ============================================================

print("\n" + "=" * 70)
print("NORMALIZATION RANGE VERIFICATION")
print("=" * 70)


for column in numerical_columns:

    minimum = preprocessed_df[column].min()

    maximum = preprocessed_df[column].max()

    print(
        f"{column:<30} "
        f"Min = {minimum:.4f} | "
        f"Max = {maximum:.4f}"
    )


# ============================================================
# 8. DEFINE CATEGORICAL INPUT FEATURES
# ============================================================

categorical_columns = [

    "behavior",

    "gender",

    "device"

]


print("\n" + "=" * 70)
print("CATEGORICAL FEATURES")
print("=" * 70)


for column in categorical_columns:

    print(
        f"\n{column}:"
    )

    print(
        preprocessed_df[
            column
        ].unique()
    )


# ============================================================
# 9. APPLY ONE-HOT ENCODING
# ============================================================

preprocessed_df = pd.get_dummies(

    preprocessed_df,

    columns=categorical_columns,

    dtype=int

)


# ============================================================
# 10. DISPLAY ENCODED FEATURES
# ============================================================

print("\n" + "=" * 70)
print("ONE-HOT ENCODED FEATURES")
print("=" * 70)


encoded_columns = [

    column

    for column in preprocessed_df.columns

    if any(

        column.startswith(
            category + "_"
        )

        for category in categorical_columns

    )

]


for column in encoded_columns:

    print("-", column)


# ============================================================
# 11. ENCODE TARGET COLUMN
# ============================================================

target_mapping = {

    "Not Engaged": 0,

    "Engaged": 1

}


preprocessed_df[
    "engagement"
] = preprocessed_df[
    "engagement"
].map(target_mapping)


# ============================================================
# 12. VERIFY TARGET ENCODING
# ============================================================

print("\n" + "=" * 70)
print("TARGET ENCODING")
print("=" * 70)


print(

    preprocessed_df[
        "engagement"
    ].value_counts()

)


print("\nTarget Mapping:")

print("Not Engaged -> 0")

print("Engaged     -> 1")


# ============================================================
# 13. DISPLAY FINAL COLUMNS
# ============================================================

print("\n" + "=" * 70)
print("PREPROCESSED DATASET COLUMNS")
print("=" * 70)


for number, column in enumerate(

    preprocessed_df.columns,

    start=1

):

    print(
        f"{number}. {column}"
    )


# ============================================================
# 14. DISPLAY FIRST FIVE RECORDS
# ============================================================

print("\n" + "=" * 70)
print("FIRST 5 PREPROCESSED RECORDS")
print("=" * 70)


print(

    preprocessed_df.head()

)


# ============================================================
# 15. CHECK MISSING VALUES
# ============================================================

print("\n" + "=" * 70)
print("MISSING VALUE VERIFICATION")
print("=" * 70)


print(

    "Total Missing Values:",

    preprocessed_df
    .isnull()
    .sum()
    .sum()

)


# ============================================================
# 16. CHECK FINAL DATASET SHAPE
# ============================================================

print("\n" + "=" * 70)
print("FINAL PREPROCESSING SUMMARY")
print("=" * 70)


print(

    "Dataset Shape Before Preprocessing:",

    merged_df.shape

)


print(

    "Dataset Shape After Preprocessing :",

    preprocessed_df.shape

)


print(

    "Numerical Features Normalized     :",

    len(numerical_columns)

)


print(

    "Categorical Features Encoded      :",

    len(categorical_columns)

)


print(

    "One-Hot Features Created          :",

    len(encoded_columns)

)


print(

    "Target Column                     : engagement"

)


print("\nSTEP 3 COMPLETED SUCCESSFULLY")
```

```python
# ============================================================
# STEP 4: IMAGE PREPROCESSING
# IMAGE MATCHING + RESIZING + NORMALIZATION + AUGMENTATION
# ============================================================

import os
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.preprocessing.image import (
    load_img,
    img_to_array
)

from tensorflow.keras.preprocessing.image import ImageDataGenerator


# ============================================================
# 1. DEFINE IMAGE FOLDER PATH
# ============================================================

IMAGE_FOLDER_PATH = r"C:\Users\SIRPI\Downloads\New folder (12)\Images"


# ============================================================
# 2. FIND ALL IMAGE FILES RECURSIVELY
# ============================================================

image_path_dictionary = {}

valid_extensions = (
    ".jpg",
    ".jpeg",
    ".png",
    ".bmp"
)

for root, directories, files in os.walk(IMAGE_FOLDER_PATH):

    for file in files:

        if file.lower().endswith(valid_extensions):

            image_path_dictionary[file] = os.path.join(
                root,
                file
            )


print("=" * 70)
print("STEP 4: IMAGE PREPROCESSING")
print("=" * 70)

print("\nTotal Actual Images Found:",
      len(image_path_dictionary))


print("\nFirst 10 Image Files:")

for i, (name, path) in enumerate(
    image_path_dictionary.items()
):

    if i == 10:
        break

    print(name, "->", path)


# ============================================================
# 3. MATCH DATASET IMAGE NAMES WITH ACTUAL IMAGES
# ============================================================

preprocessed_df["image_path"] = (
    preprocessed_df["image_name"]
    .map(image_path_dictionary)
)


print("\n" + "=" * 70)
print("IMAGE MATCHING RESULT")
print("=" * 70)

print(
    "Total Dataset Records :",
    len(preprocessed_df)
)

print(
    "Matched Images        :",
    preprocessed_df["image_path"].notna().sum()
)

print(
    "Missing Images        :",
    preprocessed_df["image_path"].isna().sum()
)


# ============================================================
# 4. DISPLAY MISSING IMAGE NAMES
# ============================================================

missing_images = (

    preprocessed_df.loc[
        preprocessed_df["image_path"].isna(),
        "image_name"
    ]

    .unique()

)


print("\nNumber of Unique Missing Images:",
      len(missing_images))


if len(missing_images) > 0:

    print("\nFirst 20 Missing Image Names:")

    print(missing_images[:20])


# ============================================================
# 5. KEEP ONLY RECORDS WITH AVAILABLE IMAGES
# ============================================================

image_df = (

    preprocessed_df
    .dropna(subset=["image_path"])
    .reset_index(drop=True)

)


print("\nDataset Shape After Image Matching:",
      image_df.shape)


# ============================================================
# 6. DEFINE IMAGE DIMENSIONS
# ============================================================

IMG_HEIGHT = 224

IMG_WIDTH = 224

IMG_CHANNELS = 3


# ============================================================
# 7. FUNCTION FOR IMAGE LOADING AND RESIZING
# ============================================================

def preprocess_image(image_path):

    try:

        # Load RGB image and resize to 224 x 224
        image = load_img(

            image_path,

            target_size=(
                IMG_HEIGHT,
                IMG_WIDTH
            ),

            color_mode="rgb"

        )


        # Convert image into numerical array
        image = img_to_array(image)


        # Convert datatype
        image = image.astype("float32")


        # Pixel normalization [0,255] -> [0,1]
        image = image / 255.0


        return image


    except Exception as error:

        print(
            "Error processing:",
            image_path,
            error
        )

        return None


# ============================================================
# 8. LOAD AND PREPROCESS ALL IMAGES
# ============================================================

processed_images = []

valid_indices = []


print("\n" + "=" * 70)
print("LOADING AND PREPROCESSING IMAGES")
print("=" * 70)


for index, image_path in enumerate(
    image_df["image_path"]
):

    processed_image = preprocess_image(
        image_path
    )


    if processed_image is not None:

        processed_images.append(
            processed_image
        )

        valid_indices.append(
            index
        )


# ============================================================
# 9. CONVERT IMAGE LIST INTO NUMPY ARRAY
# ============================================================

X_images = np.array(

    processed_images,

    dtype=np.float32

)


# Keep only successfully processed records
image_df = (

    image_df
    .iloc[valid_indices]
    .reset_index(drop=True)

)


print("\n" + "=" * 70)
print("IMAGE ARRAY DETAILS")
print("=" * 70)


print(
    "Image Array Shape:",
    X_images.shape
)


print(
    "Image Data Type:",
    X_images.dtype
)


print(
    "Minimum Pixel Value:",
    X_images.min()
)


print(
    "Maximum Pixel Value:",
    X_images.max()
)


# ============================================================
# 10. DEFINE IMAGE AUGMENTATION
# ============================================================

image_augmentation = ImageDataGenerator(

    rotation_range=15,

    horizontal_flip=True,

    zoom_range=0.1,

    width_shift_range=0.1,

    height_shift_range=0.1,

    fill_mode="nearest"

)


print("\n" + "=" * 70)
print("IMAGE AUGMENTATION SETTINGS")
print("=" * 70)


print("Random Rotation     : ±15 degrees")

print("Horizontal Flipping : Enabled")

print("Random Zoom         : Approximately 0.9–1.1")

print("Width Shift         : 10%")

print("Height Shift        : 10%")


# ============================================================
# 11. CREATE AUGMENTATION GENERATOR
# ============================================================

augmentation_generator = image_augmentation.flow(

    X_images,

    batch_size=32,

    shuffle=True

)


# ============================================================
# 12. GENERATE ONE AUGMENTED BATCH FOR VERIFICATION
# ============================================================

augmented_batch = next(
    augmentation_generator
)


print("\n" + "=" * 70)
print("AUGMENTATION VERIFICATION")
print("=" * 70)


print(
    "Original Image Array Shape:",
    X_images.shape
)


print(
    "Augmented Batch Shape:",
    augmented_batch.shape
)


print(
    "Augmented Minimum Value:",
    augmented_batch.min()
)


print(
    "Augmented Maximum Value:",
    augmented_batch.max()
)


# ============================================================
# 13. EXTRACT TARGET LABELS
# ============================================================

y = image_df[
    "engagement"
].values.astype("int32")


print("\n" + "=" * 70)
print("TARGET ARRAY")
print("=" * 70)


print(
    "Target Shape:",
    y.shape
)


print(
    "Target Classes:",
    np.unique(y)
)


print(
    "Class Distribution:"
)


unique_classes, class_counts = np.unique(

    y,

    return_counts=True

)


for class_value, count in zip(

    unique_classes,

    class_counts

):

    print(
        f"Class {class_value}: {count}"
    )


# ============================================================
# 14. FINAL IMAGE PREPROCESSING SUMMARY
# ============================================================

print("\n" + "=" * 70)
print("FINAL IMAGE PREPROCESSING SUMMARY")
print("=" * 70)


print(
    "Final Dataset Records :",
    len(image_df)
)


print(
    "Processed Image Shape :",
    X_images.shape
)


print(
    "Target Shape          :",
    y.shape
)


print(
    "Image Height          :",
    IMG_HEIGHT
)


print(
    "Image Width           :",
    IMG_WIDTH
)


print(
    "Image Channels        :",
    IMG_CHANNELS
)


print(
    "Pixel Value Range     : [0, 1]"
)


print(
    "Augmentation          : Configured"
)


print("\nSTEP 4 COMPLETED SUCCESSFULLY")
```

```python
# ============================================================
# STEP 10: TRANSFORMER ENCODER + BiLSTM MODEL
# ============================================================

import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Model

from tensorflow.keras.layers import (
    Input,
    Dense,
    Dropout,
    MultiHeadAttention,
    LayerNormalization,
    Add,
    Bidirectional,
    LSTM
)

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau,
    ModelCheckpoint
)

from tensorflow.keras.optimizers import Adam


# ============================================================
# 1. SET RANDOM SEEDS
# ============================================================

np.random.seed(42)

tf.random.set_seed(42)


# ============================================================
# 2. DISPLAY INPUT DATA DETAILS
# ============================================================

print("=" * 70)
print("STEP 10: TRANSFORMER ENCODER + BiLSTM MODEL")
print("=" * 70)

print("\nTraining Sequence Shape   :", X_train_sequence.shape)

print("Validation Sequence Shape :", X_validation_sequence.shape)

print("Testing Sequence Shape    :", X_test_sequence.shape)

print("\nTraining Target Shape     :", y_train.shape)

print("Validation Target Shape   :", y_validation.shape)

print("Testing Target Shape      :", y_test.shape)


# ============================================================
# 3. DEFINE MODEL PARAMETERS
# ============================================================

SEQUENCE_LENGTH = X_train_sequence.shape[1]

INPUT_DIMENSION = X_train_sequence.shape[2]


# Transformer parameters

NUM_HEADS = 4

KEY_DIMENSION = 16

FEED_FORWARD_DIMENSION = 64

TRANSFORMER_DROPOUT = 0.2


# BiLSTM parameters

BILSTM_UNITS = 64

BILSTM_DROPOUT = 0.3


# Dense layer parameters

DENSE_UNITS = 64

DENSE_DROPOUT = 0.3


# Training parameters

LEARNING_RATE = 0.001

BATCH_SIZE = 32

EPOCHS = 50


print("\n" + "=" * 70)
print("MODEL CONFIGURATION")
print("=" * 70)

print("Sequence Length          :", SEQUENCE_LENGTH)

print("Input Dimension          :", INPUT_DIMENSION)

print("Attention Heads          :", NUM_HEADS)

print("Attention Key Dimension  :", KEY_DIMENSION)

print("Feed Forward Dimension   :", FEED_FORWARD_DIMENSION)

print("BiLSTM Units             :", BILSTM_UNITS)

print("Dense Units              :", DENSE_UNITS)

print("Learning Rate            :", LEARNING_RATE)

print("Batch Size               :", BATCH_SIZE)

print("Maximum Epochs           :", EPOCHS)


# ============================================================
# 4. TRANSFORMER ENCODER FUNCTION
# ============================================================

def transformer_encoder(

    inputs,

    num_heads,

    key_dimension,

    feed_forward_dimension,

    dropout_rate

):


    # --------------------------------------------------------
    # MULTI-HEAD SELF-ATTENTION
    # --------------------------------------------------------

    attention_output = MultiHeadAttention(

        num_heads=num_heads,

        key_dim=key_dimension,

        dropout=dropout_rate,

        name="multi_head_self_attention"

    )(

        inputs,

        inputs

    )


    # --------------------------------------------------------
    # ATTENTION DROPOUT
    # --------------------------------------------------------

    attention_output = Dropout(

        dropout_rate,

        name="attention_dropout"

    )(attention_output)


    # --------------------------------------------------------
    # FIRST RESIDUAL CONNECTION
    # --------------------------------------------------------

    residual_attention = Add(

        name="attention_residual_connection"

    )([

        inputs,

        attention_output

    ])


    # --------------------------------------------------------
    # FIRST LAYER NORMALIZATION
    # --------------------------------------------------------

    attention_normalized = LayerNormalization(

        epsilon=1e-6,

        name="attention_layer_normalization"

    )(residual_attention)


    # --------------------------------------------------------
    # FEED FORWARD NETWORK
    # --------------------------------------------------------

    feed_forward_output = Dense(

        feed_forward_dimension,

        activation="relu",

        name="transformer_feed_forward_1"

    )(attention_normalized)


    feed_forward_output = Dropout(

        dropout_rate,

        name="feed_forward_dropout"

    )(feed_forward_output)


    # Project back to original input dimension
    feed_forward_output = Dense(

        inputs.shape[-1],

        name="transformer_feed_forward_2"

    )(feed_forward_output)


    # --------------------------------------------------------
    # SECOND RESIDUAL CONNECTION
    # --------------------------------------------------------

    residual_feed_forward = Add(

        name="feed_forward_residual_connection"

    )([

        attention_normalized,

        feed_forward_output

    ])


    # --------------------------------------------------------
    # SECOND LAYER NORMALIZATION
    # --------------------------------------------------------

    transformer_output = LayerNormalization(

        epsilon=1e-6,

        name="feed_forward_layer_normalization"

    )(residual_feed_forward)


    return transformer_output


# ============================================================
# 5. DEFINE INPUT LAYER
# ============================================================

model_input = Input(

    shape=(

        SEQUENCE_LENGTH,

        INPUT_DIMENSION

    ),

    name="multimodal_sequence_input"

)


# ============================================================
# 6. TRANSFORMER ENCODER
# ============================================================

transformer_output = transformer_encoder(

    model_input,

    num_heads=NUM_HEADS,

    key_dimension=KEY_DIMENSION,

    feed_forward_dimension=FEED_FORWARD_DIMENSION,

    dropout_rate=TRANSFORMER_DROPOUT

)


# ============================================================
# 7. BIDIRECTIONAL LSTM
# ============================================================

bilstm_output = Bidirectional(

    LSTM(

        units=BILSTM_UNITS,

        return_sequences=False,

        dropout=BILSTM_DROPOUT

    ),

    name="bidirectional_lstm"

)(transformer_output)


# ============================================================
# 8. DENSE CLASSIFICATION LAYER
# ============================================================

dense_output = Dense(

    units=DENSE_UNITS,

    activation="relu",

    name="classification_dense"

)(bilstm_output)


# ============================================================
# 9. DROPOUT
# ============================================================

dense_output = Dropout(

    DENSE_DROPOUT,

    name="classification_dropout"

)(dense_output)


# ============================================================
# 10. BINARY OUTPUT LAYER
# ============================================================

model_output = Dense(

    units=1,

    activation="sigmoid",

    name="engagement_prediction"

)(dense_output)


# ============================================================
# 11. CREATE FINAL MODEL
# ============================================================

model = Model(

    inputs=model_input,

    outputs=model_output,

    name="Transformer_BiLSTM_Model"

)


# ============================================================
# 12. DISPLAY MODEL ARCHITECTURE
# ============================================================

print("\n" + "=" * 70)
print("TRANSFORMER + BiLSTM MODEL ARCHITECTURE")
print("=" * 70)


model.summary()


# ============================================================
# 13. COMPILE MODEL
# ============================================================

optimizer = Adam(

    learning_rate=LEARNING_RATE

)


model.compile(

    optimizer=optimizer,

    loss="binary_crossentropy",

    metrics=[

        "accuracy",

        tf.keras.metrics.Precision(
            name="precision"
        ),

        tf.keras.metrics.Recall(
            name="recall"
        ),

        tf.keras.metrics.AUC(
            name="auc"
        )

    ]

)


print("\n" + "=" * 70)
print("MODEL COMPILATION")
print("=" * 70)

print("Optimizer : Adam")

print("Loss      : Binary Crossentropy")

print("Metrics   : Accuracy, Precision, Recall, AUC")


# ============================================================
# 14. EARLY STOPPING
# ============================================================

early_stopping = EarlyStopping(

    monitor="val_loss",

    patience=10,

    mode="min",

    restore_best_weights=True,

    verbose=1

)


# ============================================================
# 15. REDUCE LEARNING RATE
# ============================================================

reduce_learning_rate = ReduceLROnPlateau(

    monitor="val_loss",

    factor=0.5,

    patience=5,

    min_lr=1e-6,

    mode="min",

    verbose=1

)


# ============================================================
# 16. SAVE BEST MODEL
# ============================================================

model_checkpoint = ModelCheckpoint(

    filepath="best_transformer_bilstm_model.keras",

    monitor="val_loss",

    save_best_only=True,

    mode="min",

    verbose=1

)


# ============================================================
# 17. CALLBACK LIST
# ============================================================

callbacks = [

    early_stopping,

    reduce_learning_rate,

    model_checkpoint

]


# ============================================================
# 18. TRAIN MODEL
# ============================================================

print("\n" + "=" * 70)
print("STARTING MODEL TRAINING")
print("=" * 70)


history = model.fit(

    X_train_sequence,

    y_train,

    validation_data=(

        X_validation_sequence,

        y_validation

    ),

    epochs=EPOCHS,

    batch_size=BATCH_SIZE,

    callbacks=callbacks,

    verbose=1

)


# ============================================================
# 19. DISPLAY TRAINING COMPLETION
# ============================================================

print("\n" + "=" * 70)
print("MODEL TRAINING COMPLETED")
print("=" * 70)
```

```python
# ============================================================
# STEP 11: 5-FOLD STRATIFIED GROUP CROSS-VALIDATION
# TRANSFORMER + BiLSTM MODEL
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import gc

from sklearn.model_selection import StratifiedGroupKFold

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    mean_squared_error,
    mean_absolute_error,
    r2_score
)

from tensorflow.keras.models import Model

from tensorflow.keras.layers import (
    Input,
    Dense,
    Dropout,
    MultiHeadAttention,
    LayerNormalization,
    Add,
    Bidirectional,
    LSTM
)

from tensorflow.keras.optimizers import Adam

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)


# ============================================================
# 1. SET RANDOM SEEDS
# ============================================================

np.random.seed(42)

tf.random.set_seed(42)


# ============================================================
# 2. PREPARE COMPLETE DATASET
# ============================================================

X_cv = fused_features.astype("float32")

y_cv = y.astype("int32")

groups_cv = image_df[
    "image_name"
].values


print("=" * 70)
print("5-FOLD STRATIFIED GROUP CROSS-VALIDATION")
print("=" * 70)

print("\nFeature Shape :", X_cv.shape)

print("Target Shape  :", y_cv.shape)

print("Groups Shape  :", groups_cv.shape)

print(
    "Unique Groups :",
    len(np.unique(groups_cv))
)


# ============================================================
# 3. RESHAPE COMPLETE FEATURE DATA
# ============================================================

SEQUENCE_LENGTH = 16

FEATURES_PER_STEP = 16


X_cv_sequence = X_cv.reshape(

    X_cv.shape[0],

    SEQUENCE_LENGTH,

    FEATURES_PER_STEP

)


print(
    "\nSequence Feature Shape:",
    X_cv_sequence.shape
)


# ============================================================
# 4. VERIFY ALIGNMENT
# ============================================================

assert (

    X_cv_sequence.shape[0]

    ==

    y_cv.shape[0]

    ==

    groups_cv.shape[0]

)


print("\nFeature-Target-Group Alignment: CORRECT")


# ============================================================
# 5. MODEL BUILDING FUNCTION
# ============================================================

def build_transformer_bilstm_model():


    # --------------------------------------------------------
    # INPUT LAYER
    # --------------------------------------------------------

    inputs = Input(

        shape=(

            SEQUENCE_LENGTH,

            FEATURES_PER_STEP

        )

    )


    # --------------------------------------------------------
    # MULTI-HEAD SELF-ATTENTION
    # --------------------------------------------------------

    attention_output = MultiHeadAttention(

        num_heads=4,

        key_dim=16,

        dropout=0.2

    )(

        inputs,

        inputs

    )


    # --------------------------------------------------------
    # ATTENTION DROPOUT
    # --------------------------------------------------------

    attention_output = Dropout(

        0.2

    )(attention_output)


    # --------------------------------------------------------
    # FIRST RESIDUAL CONNECTION
    # --------------------------------------------------------

    x = Add()(

        [

            inputs,

            attention_output

        ]

    )


    # --------------------------------------------------------
    # FIRST LAYER NORMALIZATION
    # --------------------------------------------------------

    x = LayerNormalization(

        epsilon=1e-6

    )(x)


    # --------------------------------------------------------
    # TRANSFORMER FEED FORWARD NETWORK
    # --------------------------------------------------------

    feed_forward = Dense(

        64,

        activation="relu"

    )(x)


    feed_forward = Dropout(

        0.2

    )(feed_forward)


    feed_forward = Dense(

        FEATURES_PER_STEP

    )(feed_forward)


    # --------------------------------------------------------
    # SECOND RESIDUAL CONNECTION
    # --------------------------------------------------------

    x = Add()(

        [

            x,

            feed_forward

        ]

    )


    # --------------------------------------------------------
    # SECOND LAYER NORMALIZATION
    # --------------------------------------------------------

    x = LayerNormalization(

        epsilon=1e-6

    )(x)


    # --------------------------------------------------------
    # BIDIRECTIONAL LSTM
    # --------------------------------------------------------

    x = Bidirectional(

        LSTM(

            units=64,

            return_sequences=False,

            dropout=0.3

        )

    )(x)


    # --------------------------------------------------------
    # DENSE LAYER
    # --------------------------------------------------------

    x = Dense(

        64,

        activation="relu"

    )(x)


    # --------------------------------------------------------
    # DROPOUT
    # --------------------------------------------------------

    x = Dropout(

        0.3

    )(x)


    # --------------------------------------------------------
    # OUTPUT LAYER
    # --------------------------------------------------------

    outputs = Dense(

        1,

        activation="sigmoid"

    )(x)


    # --------------------------------------------------------
    # CREATE MODEL
    # --------------------------------------------------------

    cv_model = Model(

        inputs=inputs,

        outputs=outputs

    )


    # --------------------------------------------------------
    # COMPILE MODEL
    # --------------------------------------------------------

    cv_model.compile(

        optimizer=Adam(

            learning_rate=0.001

        ),

        loss="binary_crossentropy",

        metrics=["accuracy"]

    )


    return cv_model


# ============================================================
# 6. INITIALIZE STRATIFIED GROUP 5-FOLD CV
# ============================================================

sgkf = StratifiedGroupKFold(

    n_splits=5,

    shuffle=True,

    random_state=42

)


# ============================================================
# 7. CREATE RESULT STORAGE
# ============================================================

fold_results = []


# ============================================================
# 8. START 5-FOLD CROSS-VALIDATION
# ============================================================

print("\n" + "=" * 70)

print("STARTING 5-FOLD CROSS-VALIDATION")

print("=" * 70)


for fold, (

    train_index,

    validation_index

) in enumerate(

    sgkf.split(

        X_cv_sequence,

        y_cv,

        groups=groups_cv

    ),

    start=1

):


    print("\n\n" + "=" * 70)

    print(f"FOLD {fold} / 5")

    print("=" * 70)


    # ========================================================
    # CLEAR PREVIOUS MODEL SESSION
    # ========================================================

    tf.keras.backend.clear_session()

    gc.collect()


    # ========================================================
    # CREATE FOLD DATA
    # ========================================================

    X_fold_train = X_cv_sequence[
        train_index
    ]

    X_fold_validation = X_cv_sequence[
        validation_index
    ]


    y_fold_train = y_cv[
        train_index
    ]

    y_fold_validation = y_cv[
        validation_index
    ]


    train_groups = groups_cv[
        train_index
    ]

    validation_groups = groups_cv[
        validation_index
    ]


    # ========================================================
    # DISPLAY FOLD DETAILS
    # ========================================================

    print(
        "\nTraining Shape   :",
        X_fold_train.shape
    )

    print(
        "Validation Shape :",
        X_fold_validation.shape
    )


    print(
        "\nTraining Groups   :",
        len(np.unique(train_groups))
    )

    print(
        "Validation Groups :",
        len(np.unique(validation_groups))
    )


    # ========================================================
    # CHECK GROUP LEAKAGE
    # ========================================================

    group_overlap = (

        set(train_groups)

        &

        set(validation_groups)

    )


    print(
        "Group Overlap     :",
        len(group_overlap)
    )


    assert len(group_overlap) == 0


    # ========================================================
    # DISPLAY CLASS DISTRIBUTION
    # ========================================================

    print(
        "\nTraining Class Distribution:"
    )

    print(

        np.bincount(
            y_fold_train
        )

    )


    print(
        "Validation Class Distribution:"
    )

    print(

        np.bincount(
            y_fold_validation
        )

    )


    # ========================================================
    # BUILD NEW MODEL
    # ========================================================

    cv_model = build_transformer_bilstm_model()


    print(
        "\nNew Transformer-BiLSTM Model Created"
    )


    # ========================================================
    # DEFINE CALLBACKS
    # ========================================================

    early_stopping = EarlyStopping(

        monitor="val_loss",

        patience=8,

        restore_best_weights=True,

        mode="min",

        verbose=0

    )


    reduce_lr = ReduceLROnPlateau(

        monitor="val_loss",

        factor=0.5,

        patience=4,

        min_lr=1e-6,

        mode="min",

        verbose=0

    )


    # ========================================================
    # TRAIN MODEL
    # ========================================================

    print(
        f"\nTraining Fold {fold}..."
    )


    fold_history = cv_model.fit(

        X_fold_train,

        y_fold_train,

        validation_data=(

            X_fold_validation,

            y_fold_validation

        ),

        epochs=50,

        batch_size=32,

        callbacks=[

            early_stopping,

            reduce_lr

        ],

        verbose=0

    )


    print(
        "Epochs Trained:",
        len(fold_history.history["loss"])
    )


    # ========================================================
    # GENERATE PROBABILITY PREDICTIONS
    # ========================================================

    y_fold_probability = cv_model.predict(

        X_fold_validation,

        batch_size=32,

        verbose=0

    ).ravel()


    # ========================================================
    # CONVERT PROBABILITY INTO CLASS
    # ========================================================

    y_fold_prediction = (

        y_fold_probability >= 0.5

    ).astype("int32")


    # ========================================================
    # CALCULATE CLASSIFICATION METRICS
    # ========================================================

    fold_accuracy = accuracy_score(

        y_fold_validation,

        y_fold_prediction

    )


    fold_precision = precision_score(

        y_fold_validation,

        y_fold_prediction,

        zero_division=0

    )


    fold_recall = recall_score(

        y_fold_validation,

        y_fold_prediction,

        zero_division=0

    )


    fold_f1 = f1_score(

        y_fold_validation,

        y_fold_prediction,

        zero_division=0

    )


    fold_auc = roc_auc_score(

        y_fold_validation,

        y_fold_probability

    )


    # ========================================================
    # CALCULATE ERROR METRICS
    # ========================================================

    fold_mse = mean_squared_error(

        y_fold_validation,

        y_fold_probability

    )


    fold_mae = mean_absolute_error(

        y_fold_validation,

        y_fold_probability

    )


    fold_rmse = np.sqrt(

        fold_mse

    )


    fold_r2 = r2_score(

        y_fold_validation,

        y_fold_probability

    )


    # ========================================================
    # STORE RESULTS
    # ========================================================

    fold_results.append({

        "Fold": fold,

        "Accuracy": fold_accuracy,

        "Precision": fold_precision,

        "Recall": fold_recall,

        "F1-Score": fold_f1,

        "AUC": fold_auc,

        "MSE": fold_mse,

        "MAE": fold_mae,

        "RMSE": fold_rmse,

        "R2": fold_r2

    })


    # ========================================================
    # DISPLAY FOLD RESULTS
    # ========================================================

    print("\n" + "-" * 70)

    print(f"FOLD {fold} RESULTS")

    print("-" * 70)


    print(
        f"Accuracy  : {fold_accuracy:.6f}"
    )

    print(
        f"Precision : {fold_precision:.6f}"
    )

    print(
        f"Recall    : {fold_recall:.6f}"
    )

    print(
        f"F1-Score  : {fold_f1:.6f}"
    )

    print(
        f"AUC       : {fold_auc:.6f}"
    )

    print(
        f"MSE       : {fold_mse:.6f}"
    )

    print(
        f"MAE       : {fold_mae:.6f}"
    )

    print(
        f"RMSE      : {fold_rmse:.6f}"
    )

    print(
        f"R2        : {fold_r2:.6f}"
    )


# ============================================================
# 9. CONVERT RESULTS TO DATAFRAME
# ============================================================

cv_results_df = pd.DataFrame(

    fold_results

)


# ============================================================
# 10. DISPLAY ALL FOLD RESULTS
# ============================================================

print("\n\n" + "=" * 90)

print("5-FOLD CROSS-VALIDATION RESULTS")

print("=" * 90)


print(

    cv_results_df.to_string(

        index=False

    )

)


# ============================================================
# 11. CALCULATE MEAN RESULTS
# ============================================================

mean_results = (

    cv_results_df

    .drop(columns=["Fold"])

    .mean()

)


# ============================================================
# 12. CALCULATE STANDARD DEVIATION
# ============================================================

std_results = (

    cv_results_df

    .drop(columns=["Fold"])

    .std()

)


# ============================================================
# 13. DISPLAY MEAN RESULTS
# ============================================================

print("\n" + "=" * 70)

print("AVERAGE 5-FOLD CROSS-VALIDATION RESULTS")

print("=" * 70)


for metric in mean_results.index:

    print(

        f"{metric:<12}: "

        f"{mean_results[metric]:.6f} "

        f"± "

        f"{std_results[metric]:.6f}"

    )


# ============================================================
# 14. SAVE RESULTS
# ============================================================

cv_results_df.to_csv(

    "5_fold_cross_validation_results.csv",

    index=False

)


# ============================================================
# 15. FINAL SUMMARY
# ============================================================

print("\n" + "=" * 70)

print("FINAL 5-FOLD CROSS-VALIDATION SUMMARY")

print("=" * 70)


print(

    f"Average Accuracy  : "

    f"{mean_results['Accuracy']:.6f}"

)


print(

    f"Average Precision : "

    f"{mean_results['Precision']:.6f}"

)


print(

    f"Average Recall    : "

    f"{mean_results['Recall']:.6f}"

)


print(

    f"Average F1-Score  : "

    f"{mean_results['F1-Score']:.6f}"

)


print(

    f"Average AUC       : "

    f"{mean_results['AUC']:.6f}"

)


print(

    f"Average MSE       : "

    f"{mean_results['MSE']:.6f}"

)


print(

    f"Average MAE       : "

    f"{mean_results['MAE']:.6f}"

)


print(

    f"Average RMSE      : "

    f"{mean_results['RMSE']:.6f}"

)


print(

    f"Average R2        : "

    f"{mean_results['R2']:.6f}"

)


print("\nResults Saved As:")

print("5_fold_cross_validation_results.csv")


print("\n5-FOLD CROSS-VALIDATION COMPLETED SUCCESSFULLY")
```

```python
# ============================================================
# ABLATION STUDY
# TO-TBi-STMNet
#
# CONFIGURATIONS:
#
# 1. Preprocessing Only
# 2. + Multimodal Fusion
# 3. + BiLSTM
# 4. + Transformer
# 5. Full Model (TO-TBi-STMNet)
# ============================================================


# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import gc
import time

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

from sklearn.model_selection import GroupShuffleSplit

from tensorflow.keras.models import Model

from tensorflow.keras.layers import (
    Input,
    Dense,
    Dropout,
    Bidirectional,
    LSTM,
    MultiHeadAttention,
    LayerNormalization,
    Add,
    GlobalAveragePooling1D
)

from tensorflow.keras.optimizers import Adam

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)


# ============================================================
# 2. RANDOM SEEDS
# ============================================================

np.random.seed(42)

tf.random.set_seed(42)


# ============================================================
# 3. TRAINING PARAMETERS
# ============================================================

EPOCHS = 50

BATCH_SIZE = 32


print("=" * 80)
print("TO-TBi-STMNet ABLATION STUDY")
print("=" * 80)


# ============================================================
# 4. PREPARE TARGET AND IMAGE GROUPS
# ============================================================

y_ablation = y.astype("int32")

groups_ablation = image_df[
    "image_name"
].values


print("\nTotal Samples :", len(y_ablation))

print(
    "Unique Groups:",
    len(np.unique(groups_ablation))
)


# ============================================================
# 5. CREATE COMMON GROUP-AWARE SPLIT
#
# IMPORTANT:
# SAME SPLIT IS USED FOR ALL ABLATION MODELS
# ============================================================

group_split_1 = GroupShuffleSplit(

    n_splits=1,

    train_size=0.70,

    random_state=42

)


train_indices, temporary_indices = next(

    group_split_1.split(

        fused_features,

        y_ablation,

        groups=groups_ablation

    )

)


# ============================================================
# 6. VALIDATION / TEST SPLIT
# ============================================================

temporary_groups = groups_ablation[
    temporary_indices
]


group_split_2 = GroupShuffleSplit(

    n_splits=1,

    train_size=0.50,

    random_state=42

)


validation_relative_indices, test_relative_indices = next(

    group_split_2.split(

        fused_features[temporary_indices],

        y_ablation[temporary_indices],

        groups=temporary_groups

    )

)


validation_indices = temporary_indices[
    validation_relative_indices
]


test_indices = temporary_indices[
    test_relative_indices
]


print("\n" + "=" * 80)
print("COMMON DATA SPLIT")
print("=" * 80)

print(
    "Training Samples   :",
    len(train_indices)
)

print(
    "Validation Samples :",
    len(validation_indices)
)

print(
    "Testing Samples    :",
    len(test_indices)
)


# ============================================================
# 7. VERIFY GROUP LEAKAGE
# ============================================================

train_groups = set(

    groups_ablation[
        train_indices
    ]

)


validation_groups = set(

    groups_ablation[
        validation_indices
    ]

)


test_groups = set(

    groups_ablation[
        test_indices
    ]

)


print("\nTrain-Validation Overlap:",

      len(
          train_groups
          &
          validation_groups
      )
)


print("Train-Test Overlap:",

      len(
          train_groups
          &
          test_groups
      )
)


print("Validation-Test Overlap:",

      len(
          validation_groups
          &
          test_groups
      )
)


assert len(train_groups & validation_groups) == 0

assert len(train_groups & test_groups) == 0

assert len(validation_groups & test_groups) == 0


print("\nGroup Leakage: NOT DETECTED")


# ============================================================
# 8. TARGET SPLITS
# ============================================================

y_train_ablation = y_ablation[
    train_indices
]


y_validation_ablation = y_ablation[
    validation_indices
]


y_test_ablation = y_ablation[
    test_indices
]


# ============================================================
# 9. PREPARE PREPROCESSING-ONLY FEATURES
#
# NO CNN
# NO DNN FEATURE EXTRACTION
# NO TEXT FEATURES
# NO MULTIMODAL FUSION
#
# USE DIRECT PREPROCESSED STRUCTURED FEATURES
# ============================================================

X_preprocessing = image_df[
    structured_feature_columns
].values.astype("float32")


X_pre_train = X_preprocessing[
    train_indices
]


X_pre_validation = X_preprocessing[
    validation_indices
]


X_pre_test = X_preprocessing[
    test_indices
]


print("\n" + "=" * 80)
print("PREPROCESSING-ONLY FEATURES")
print("=" * 80)

print(
    "Complete Shape:",
    X_preprocessing.shape
)

print(
    "Training Shape:",
    X_pre_train.shape
)

print(
    "Validation Shape:",
    X_pre_validation.shape
)

print(
    "Testing Shape:",
    X_pre_test.shape
)


# ============================================================
# 10. PREPARE MULTIMODAL FUSED FEATURES
# ============================================================

X_fusion = fused_features.astype(
    "float32"
)


X_fusion_train = X_fusion[
    train_indices
]


X_fusion_validation = X_fusion[
    validation_indices
]


X_fusion_test = X_fusion[
    test_indices
]


print("\n" + "=" * 80)
print("MULTIMODAL FUSION FEATURES")
print("=" * 80)

print(
    "Complete Shape:",
    X_fusion.shape
)

print(
    "Training Shape:",
    X_fusion_train.shape
)

print(
    "Validation Shape:",
    X_fusion_validation.shape
)

print(
    "Testing Shape:",
    X_fusion_test.shape
)


# ============================================================
# 11. RESHAPE FUSION FEATURES FOR SEQUENCE MODELS
# ============================================================

SEQUENCE_LENGTH = 16

FEATURES_PER_STEP = 16


X_sequence = X_fusion.reshape(

    X_fusion.shape[0],

    SEQUENCE_LENGTH,

    FEATURES_PER_STEP

)


X_sequence_train = X_sequence[
    train_indices
]


X_sequence_validation = X_sequence[
    validation_indices
]


X_sequence_test = X_sequence[
    test_indices
]


print("\n" + "=" * 80)
print("SEQUENCE FEATURES")
print("=" * 80)

print(
    "Complete Shape:",
    X_sequence.shape
)

print(
    "Training Shape:",
    X_sequence_train.shape
)

print(
    "Validation Shape:",
    X_sequence_validation.shape
)

print(
    "Testing Shape:",
    X_sequence_test.shape
)


# ============================================================
# 12. COMMON MODEL TRAINING FUNCTION
# ============================================================

def train_and_evaluate_model(

    model,

    X_train_data,

    X_validation_data,

    X_test_data,

    model_name,

    epochs=EPOCHS,

    batch_size=BATCH_SIZE

):


    print("\n\n" + "=" * 80)

    print(model_name)

    print("=" * 80)


    # --------------------------------------------------------
    # CALLBACKS
    # --------------------------------------------------------

    early_stopping = EarlyStopping(

        monitor="val_loss",

        patience=8,

        restore_best_weights=True,

        mode="min",

        verbose=0

    )


    reduce_lr = ReduceLROnPlateau(

        monitor="val_loss",

        factor=0.5,

        patience=4,

        min_lr=1e-6,

        mode="min",

        verbose=0

    )


    # --------------------------------------------------------
    # START TRAINING TIMER
    # --------------------------------------------------------

    start_time = time.time()


    # --------------------------------------------------------
    # TRAIN MODEL
    # --------------------------------------------------------

    history = model.fit(

        X_train_data,

        y_train_ablation,

        validation_data=(

            X_validation_data,

            y_validation_ablation

        ),

        epochs=epochs,

        batch_size=batch_size,

        callbacks=[

            early_stopping,

            reduce_lr

        ],

        verbose=0

    )


    # --------------------------------------------------------
    # TRAINING TIME
    # --------------------------------------------------------

    training_time = (

        time.time()

        -

        start_time

    )


    # --------------------------------------------------------
    # PREDICTIONS
    # --------------------------------------------------------

    probabilities = model.predict(

        X_test_data,

        batch_size=batch_size,

        verbose=0

    ).ravel()


    predictions = (

        probabilities >= 0.5

    ).astype("int32")


    # --------------------------------------------------------
    # METRICS
    # --------------------------------------------------------

    accuracy = accuracy_score(

        y_test_ablation,

        predictions

    )


    precision = precision_score(

        y_test_ablation,

        predictions,

        zero_division=0

    )


    recall = recall_score(

        y_test_ablation,

        predictions,

        zero_division=0

    )


    f1 = f1_score(

        y_test_ablation,

        predictions,

        zero_division=0

    )


    auc = roc_auc_score(

        y_test_ablation,

        probabilities

    )


    # --------------------------------------------------------
    # DISPLAY RESULTS
    # --------------------------------------------------------

    print(
        "Epochs Trained :",
        len(history.history["loss"])
    )


    print(
        f"Accuracy       : {accuracy:.6f}"
    )


    print(
        f"Precision      : {precision:.6f}"
    )


    print(
        f"Recall         : {recall:.6f}"
    )


    print(
        f"F1-Score       : {f1:.6f}"
    )


    print(
        f"AUC            : {auc:.6f}"
    )


    print(
        f"Training Time  : {training_time:.2f} seconds"
    )


    return {

        "Accuracy": accuracy,

        "Precision": precision,

        "Recall": recall,

        "F1-Score": f1,

        "AUC": auc,

        "Epochs": len(
            history.history["loss"]
        ),

        "Training Time": training_time

    }


# ============================================================
# 13. RESULT STORAGE
# ============================================================

ablation_results = []


# ============================================================
# CONFIGURATION 1
#
# PREPROCESSING ONLY
#
# PREPROCESSING     ✓
# MULTIMODAL FUSION ✗
# BiLSTM            ✗
# TRANSFORMER       ✗
# TO OPTIMIZATION   ✗
# ============================================================

tf.keras.backend.clear_session()

gc.collect()


input_pre = Input(

    shape=(X_pre_train.shape[1],)

)


x = Dense(

    64,

    activation="relu"

)(input_pre)


x = Dropout(

    0.3

)(x)


x = Dense(

    32,

    activation="relu"

)(x)


output_pre = Dense(

    1,

    activation="sigmoid"

)(x)


model_preprocessing = Model(

    input_pre,

    output_pre

)


model_preprocessing.compile(

    optimizer=Adam(

        learning_rate=0.001

    ),

    loss="binary_crossentropy",

    metrics=["accuracy"]

)


result = train_and_evaluate_model(

    model_preprocessing,

    X_pre_train,

    X_pre_validation,

    X_pre_test,

    "CONFIGURATION 1: PREPROCESSING ONLY"

)


ablation_results.append({

    "Ablation Configuration":
        "Preprocessing Only",

    "Component Removed / Added":
        "Only preprocessing; no feature fusion",

    "Data Preprocessing": "✓",

    "Multimodal Fusion": "✗",

    "BiLSTM": "✗",

    "Transformer": "✗",

    "TO Optimization": "✗",

    **result

})


# ============================================================
# CONFIGURATION 2
#
# MULTIMODAL FUSION
#
# PREPROCESSING     ✓
# MULTIMODAL FUSION ✓
# BiLSTM            ✗
# TRANSFORMER       ✗
# TO OPTIMIZATION   ✗
# ============================================================

tf.keras.backend.clear_session()

gc.collect()


input_fusion = Input(

    shape=(X_fusion_train.shape[1],)

)


x = Dense(

    128,

    activation="relu"

)(input_fusion)


x = Dropout(

    0.3

)(x)


x = Dense(

    64,

    activation="relu"

)(x)


output_fusion = Dense(

    1,

    activation="sigmoid"

)(x)


model_fusion = Model(

    input_fusion,

    output_fusion

)


model_fusion.compile(

    optimizer=Adam(

        learning_rate=0.001

    ),

    loss="binary_crossentropy",

    metrics=["accuracy"]

)


result = train_and_evaluate_model(

    model_fusion,

    X_fusion_train,

    X_fusion_validation,

    X_fusion_test,

    "CONFIGURATION 2: + MULTIMODAL FUSION"

)


ablation_results.append({

    "Ablation Configuration":
        "+ Multimodal Fusion",

    "Component Removed / Added":
        "Image, structured, and text feature fusion",

    "Data Preprocessing": "✓",

    "Multimodal Fusion": "✓",

    "BiLSTM": "✗",

    "Transformer": "✗",

    "TO Optimization": "✗",

    **result

})


# ============================================================
# CONFIGURATION 3
#
# MULTIMODAL FUSION + BiLSTM
#
# PREPROCESSING     ✓
# MULTIMODAL FUSION ✓
# BiLSTM            ✓
# TRANSFORMER       ✗
# TO OPTIMIZATION   ✗
# ============================================================

tf.keras.backend.clear_session()

gc.collect()


input_bilstm = Input(

    shape=(

        SEQUENCE_LENGTH,

        FEATURES_PER_STEP

    )

)


x = Bidirectional(

    LSTM(

        64,

        dropout=0.3

    )

)(input_bilstm)


x = Dense(

    64,

    activation="relu"

)(x)


x = Dropout(

    0.3

)(x)


output_bilstm = Dense(

    1,

    activation="sigmoid"

)(x)


model_bilstm = Model(

    input_bilstm,

    output_bilstm

)


model_bilstm.compile(

    optimizer=Adam(

        learning_rate=0.001

    ),

    loss="binary_crossentropy",

    metrics=["accuracy"]

)


result = train_and_evaluate_model(

    model_bilstm,

    X_sequence_train,

    X_sequence_validation,

    X_sequence_test,

    "CONFIGURATION 3: + BiLSTM"

)


ablation_results.append({

    "Ablation Configuration":
        "+ BiLSTM",

    "Component Removed / Added":
        "BiLSTM added; Transformer and TO excluded",

    "Data Preprocessing": "✓",

    "Multimodal Fusion": "✓",

    "BiLSTM": "✓",

    "Transformer": "✗",

    "TO Optimization": "✗",

    **result

})


# ============================================================
# FUNCTION:
# TRANSFORMER + BiLSTM MODEL
# ============================================================

def create_transformer_bilstm(

    num_heads=4,

    key_dim=16,

    feed_forward_dim=64,

    bilstm_units=64,

    dropout_rate=0.3,

    learning_rate=0.001

):


    inputs = Input(

        shape=(

            SEQUENCE_LENGTH,

            FEATURES_PER_STEP

        )

    )


    # --------------------------------------------------------
    # MULTI-HEAD ATTENTION
    # --------------------------------------------------------

    attention = MultiHeadAttention(

        num_heads=num_heads,

        key_dim=key_dim,

        dropout=0.2

    )(

        inputs,

        inputs

    )


    # --------------------------------------------------------
    # RESIDUAL CONNECTION
    # --------------------------------------------------------

    x = Add()(

        [

            inputs,

            attention

        ]

    )


    # --------------------------------------------------------
    # LAYER NORMALIZATION
    # --------------------------------------------------------

    x = LayerNormalization(

        epsilon=1e-6

    )(x)


    # --------------------------------------------------------
    # FEED FORWARD NETWORK
    # --------------------------------------------------------

    feed_forward = Dense(

        feed_forward_dim,

        activation="relu"

    )(x)


    feed_forward = Dropout(

        0.2

    )(feed_forward)


    feed_forward = Dense(

        FEATURES_PER_STEP

    )(feed_forward)


    # --------------------------------------------------------
    # SECOND RESIDUAL CONNECTION
    # --------------------------------------------------------

    x = Add()(

        [

            x,

            feed_forward

        ]

    )


    # --------------------------------------------------------
    # SECOND NORMALIZATION
    # --------------------------------------------------------

    x = LayerNormalization(

        epsilon=1e-6

    )(x)


    # --------------------------------------------------------
    # BiLSTM
    # --------------------------------------------------------

    x = Bidirectional(

        LSTM(

            bilstm_units,

            dropout=dropout_rate

        )

    )(x)


    # --------------------------------------------------------
    # DENSE
    # --------------------------------------------------------

    x = Dense(

        64,

        activation="relu"

    )(x)


    x = Dropout(

        dropout_rate

    )(x)


    # --------------------------------------------------------
    # OUTPUT
    # --------------------------------------------------------

    outputs = Dense(

        1,

        activation="sigmoid"

    )(x)


    model = Model(

        inputs,

        outputs

    )


    model.compile(

        optimizer=Adam(

            learning_rate=learning_rate

        ),

        loss="binary_crossentropy",

        metrics=["accuracy"]

    )


    return model


# ============================================================
# CONFIGURATION 4
#
# TRANSFORMER + BiLSTM
#
# PREPROCESSING     ✓
# MULTIMODAL FUSION ✓
# BiLSTM            ✓
# TRANSFORMER       ✓
# TO OPTIMIZATION   ✗
# ============================================================

tf.keras.backend.clear_session()

gc.collect()


model_transformer = create_transformer_bilstm(

    num_heads=4,

    key_dim=16,

    feed_forward_dim=64,

    bilstm_units=64,

    dropout_rate=0.3,

    learning_rate=0.001

)


result = train_and_evaluate_model(

    model_transformer,

    X_sequence_train,

    X_sequence_validation,

    X_sequence_test,

    "CONFIGURATION 4: + TRANSFORMER"

)


ablation_results.append({

    "Ablation Configuration":
        "+ Transformer",

    "Component Removed / Added":
        "Transformer added; TO optimization excluded",

    "Data Preprocessing": "✓",

    "Multimodal Fusion": "✓",

    "BiLSTM": "✓",

    "Transformer": "✓",

    "TO Optimization": "✗",

    **result

})


# ============================================================
# CONFIGURATION 5
#
# FULL MODEL
#
# TO-STYLE OPTIMIZATION
#
# IMPORTANT:
# THIS SECTION SEARCHES A SMALL HYPERPARAMETER SPACE
# USING THE VALIDATION SET ONLY.
# ============================================================


# ============================================================
# 14. DEFINE TO SEARCH SPACE
# ============================================================

to_candidates = [

    {
        "num_heads": 2,
        "key_dim": 8,
        "feed_forward_dim": 32,
        "bilstm_units": 32,
        "dropout_rate": 0.2,
        "learning_rate": 0.001
    },

    {
        "num_heads": 4,
        "key_dim": 16,
        "feed_forward_dim": 64,
        "bilstm_units": 64,
        "dropout_rate": 0.3,
        "learning_rate": 0.001
    },

    {
        "num_heads": 8,
        "key_dim": 16,
        "feed_forward_dim": 128,
        "bilstm_units": 64,
        "dropout_rate": 0.3,
        "learning_rate": 0.0005
    },

    {
        "num_heads": 4,
        "key_dim": 32,
        "feed_forward_dim": 128,
        "bilstm_units": 128,
        "dropout_rate": 0.4,
        "learning_rate": 0.0005
    }

]


# ============================================================
# 15. TO OPTIMIZATION
# ============================================================

print("\n\n" + "=" * 80)

print("TO HYPERPARAMETER OPTIMIZATION")

print("=" * 80)


best_validation_accuracy = -1

best_parameters = None


for candidate_number, parameters in enumerate(

    to_candidates,

    start=1

):


    print("\n" + "-" * 80)

    print(

        f"TO CANDIDATE {candidate_number}"

    )

    print("-" * 80)


    print(parameters)


    tf.keras.backend.clear_session()

    gc.collect()


    candidate_model = create_transformer_bilstm(

        **parameters

    )


    candidate_early_stopping = EarlyStopping(

        monitor="val_loss",

        patience=5,

        restore_best_weights=True,

        verbose=0

    )


    candidate_model.fit(

        X_sequence_train,

        y_train_ablation,

        validation_data=(

            X_sequence_validation,

            y_validation_ablation

        ),

        epochs=30,

        batch_size=32,

        callbacks=[

            candidate_early_stopping

        ],

        verbose=0

    )


    validation_probabilities = candidate_model.predict(

        X_sequence_validation,

        verbose=0

    ).ravel()


    validation_predictions = (

        validation_probabilities >= 0.5

    ).astype("int32")


    validation_accuracy = accuracy_score(

        y_validation_ablation,

        validation_predictions

    )


    print(

        f"Validation Accuracy: "
        f"{validation_accuracy:.6f}"

    )


    if validation_accuracy > best_validation_accuracy:


        best_validation_accuracy = (

            validation_accuracy

        )


        best_parameters = (

            parameters.copy()

        )


# ============================================================
# 16. DISPLAY BEST TO PARAMETERS
# ============================================================

print("\n" + "=" * 80)

print("BEST TO OPTIMIZATION RESULT")

print("=" * 80)


print(

    "Best Validation Accuracy:",

    best_validation_accuracy

)


print(

    "Best Parameters:",

    best_parameters

)


# ============================================================
# 17. TRAIN FULL OPTIMIZED MODEL
# ============================================================

tf.keras.backend.clear_session()

gc.collect()


full_model = create_transformer_bilstm(

    **best_parameters

)


result = train_and_evaluate_model(

    full_model,

    X_sequence_train,

    X_sequence_validation,

    X_sequence_test,

    "CONFIGURATION 5: FULL MODEL (TO-TBi-STMNet)"

)


ablation_results.append({

    "Ablation Configuration":
        "Full Model (TO-TBi-STMNet)",

    "Component Removed / Added":
        "Complete optimized method",

    "Data Preprocessing": "✓",

    "Multimodal Fusion": "✓",

    "BiLSTM": "✓",

    "Transformer": "✓",

    "TO Optimization": "✓",

    **result

})


# ============================================================
# 18. CREATE FINAL ABLATION TABLE
# ============================================================

ablation_results_df = pd.DataFrame(

    ablation_results

)


# ============================================================
# 19. CONVERT ACCURACY INTO PERCENTAGE
# ============================================================

ablation_results_df[

    "Accuracy (%)"

] = (

    ablation_results_df["Accuracy"]

    *

    100

)


# ============================================================
# 20. DEFINE FINAL TABLE COLUMNS
# ============================================================

final_columns = [

    "Ablation Configuration",

    "Component Removed / Added",

    "Data Preprocessing",

    "Multimodal Fusion",

    "BiLSTM",

    "Transformer",

    "TO Optimization",

    "Accuracy (%)"

]


final_ablation_table = (

    ablation_results_df[
        final_columns
    ]

)


# ============================================================
# 21. DISPLAY FINAL ABLATION TABLE
# ============================================================

print("\n\n" + "=" * 120)

print("FINAL ABLATION STUDY RESULTS")

print("=" * 120)


print(

    final_ablation_table.to_string(

        index=False,

        formatters={

            "Accuracy (%)":

                lambda x: f"{x:.2f}"

        }

    )

)


# ============================================================
# 22. DISPLAY DETAILED METRICS
# ============================================================

print("\n\n" + "=" * 120)

print("DETAILED ABLATION METRICS")

print("=" * 120)


detailed_columns = [

    "Ablation Configuration",

    "Accuracy",

    "Precision",

    "Recall",

    "F1-Score",

    "AUC",

    "Epochs",

    "Training Time"

]


print(

    ablation_results_df[

        detailed_columns

    ].to_string(

        index=False

    )

)


# ============================================================
# 23. SAVE FINAL RESULTS
# ============================================================

final_ablation_table.to_csv(

    "ablation_study_results.csv",

    index=False

)


ablation_results_df.to_csv(

    "detailed_ablation_study_results.csv",

    index=False

)


# ============================================================
# 24. SAVE BEST TO PARAMETERS
# ============================================================

with open(

    "best_to_parameters.txt",

    "w"

) as file:


    file.write(

        str(
            best_parameters
        )

    )


# ============================================================
# 25. FINAL SUMMARY
# ============================================================

print("\n" + "=" * 80)

print("ABLATION STUDY SUMMARY")

print("=" * 80)


for _, row in final_ablation_table.iterrows():


    print(

        f"{row['Ablation Configuration']:<35}"

        f": "

        f"{row['Accuracy (%)']:.2f}%"

    )


print("\nSaved Files:")

print("1. ablation_study_results.csv")

print("2. detailed_ablation_study_results.csv")

print("3. best_to_parameters.txt")


print("\nABLATION STUDY COMPLETED SUCCESSFULLY")
```

```python
# ============================================================
# STATISTICAL SIGNIFICANCE ANALYSIS
# BASELINE BiLSTM vs TO-TBi-STMNet
#
# METRICS:
# Accuracy
# Precision
# Recall
# F1-Score
#
# OUTPUT:
# Mean (%)
# 95% Confidence Interval
# t-value
# p-value
# Significance
# ============================================================


# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import gc

from scipy.stats import ttest_rel, t

from sklearn.model_selection import StratifiedGroupKFold

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

from tensorflow.keras.models import Model

from tensorflow.keras.layers import (
    Input,
    Dense,
    Dropout,
    Bidirectional,
    LSTM,
    MultiHeadAttention,
    LayerNormalization,
    Add
)

from tensorflow.keras.optimizers import Adam

from tensorflow.keras.callbacks import EarlyStopping


# ============================================================
# 2. RANDOM SEED
# ============================================================

np.random.seed(42)

tf.random.set_seed(42)


# ============================================================
# 3. PREPARE DATA
# ============================================================

X_statistical = fused_features.astype("float32")

y_statistical = y.astype("int32")

groups_statistical = image_df[
    "image_name"
].values


# ============================================================
# 4. RESHAPE FEATURES
# ============================================================

SEQUENCE_LENGTH = 16

FEATURES_PER_STEP = 16


X_statistical = X_statistical.reshape(

    X_statistical.shape[0],

    SEQUENCE_LENGTH,

    FEATURES_PER_STEP

)


print("=" * 80)

print("STATISTICAL SIGNIFICANCE ANALYSIS")

print("=" * 80)


print(
    "\nFeature Shape:",
    X_statistical.shape
)

print(
    "Target Shape:",
    y_statistical.shape
)

print(
    "Unique Groups:",
    len(
        np.unique(
            groups_statistical
        )
    )
)


# ============================================================
# 5. BASELINE BiLSTM MODEL
# ============================================================

def create_baseline_bilstm():


    inputs = Input(

        shape=(

            SEQUENCE_LENGTH,

            FEATURES_PER_STEP

        )

    )


    x = Bidirectional(

        LSTM(

            units=64,

            dropout=0.3

        )

    )(inputs)


    x = Dense(

        units=64,

        activation="relu"

    )(x)


    x = Dropout(

        rate=0.3

    )(x)


    outputs = Dense(

        units=1,

        activation="sigmoid"

    )(x)


    model = Model(

        inputs=inputs,

        outputs=outputs

    )


    model.compile(

        optimizer=Adam(

            learning_rate=0.001

        ),

        loss="binary_crossentropy",

        metrics=["accuracy"]

    )


    return model


# ============================================================
# 6. FULL TO-TBi-STMNet MODEL
#
# Replace these parameters with your actual
# best TO-optimized parameters if available.
# ============================================================

def create_full_model():


    inputs = Input(

        shape=(

            SEQUENCE_LENGTH,

            FEATURES_PER_STEP

        )

    )


    # --------------------------------------------------------
    # MULTI-HEAD SELF ATTENTION
    # --------------------------------------------------------

    attention = MultiHeadAttention(

        num_heads=4,

        key_dim=16,

        dropout=0.2

    )(

        inputs,

        inputs

    )


    # --------------------------------------------------------
    # FIRST RESIDUAL CONNECTION
    # --------------------------------------------------------

    x = Add()(

        [

            inputs,

            attention

        ]

    )


    # --------------------------------------------------------
    # FIRST NORMALIZATION
    # --------------------------------------------------------

    x = LayerNormalization(

        epsilon=1e-6

    )(x)


    # --------------------------------------------------------
    # FEED FORWARD NETWORK
    # --------------------------------------------------------

    feed_forward = Dense(

        64,

        activation="relu"

    )(x)


    feed_forward = Dropout(

        0.2

    )(feed_forward)


    feed_forward = Dense(

        FEATURES_PER_STEP

    )(feed_forward)


    # --------------------------------------------------------
    # SECOND RESIDUAL CONNECTION
    # --------------------------------------------------------

    x = Add()(

        [

            x,

            feed_forward

        ]

    )


    # --------------------------------------------------------
    # SECOND NORMALIZATION
    # --------------------------------------------------------

    x = LayerNormalization(

        epsilon=1e-6

    )(x)


    # --------------------------------------------------------
    # BiLSTM
    # --------------------------------------------------------

    x = Bidirectional(

        LSTM(

            units=64,

            dropout=0.3

        )

    )(x)


    # --------------------------------------------------------
    # DENSE LAYER
    # --------------------------------------------------------

    x = Dense(

        units=64,

        activation="relu"

    )(x)


    x = Dropout(

        rate=0.3

    )(x)


    # --------------------------------------------------------
    # OUTPUT
    # --------------------------------------------------------

    outputs = Dense(

        units=1,

        activation="sigmoid"

    )(x)


    model = Model(

        inputs=inputs,

        outputs=outputs

    )


    model.compile(

        optimizer=Adam(

            learning_rate=0.001

        ),

        loss="binary_crossentropy",

        metrics=["accuracy"]

    )


    return model


# ============================================================
# 7. INITIALIZE 5-FOLD STRATIFIED GROUP CV
# ============================================================

sgkf = StratifiedGroupKFold(

    n_splits=5,

    shuffle=True,

    random_state=42

)


# ============================================================
# 8. METRIC STORAGE
# ============================================================

baseline_scores = {

    "Accuracy": [],

    "Precision": [],

    "Recall": [],

    "F1-Score": []

}


full_model_scores = {

    "Accuracy": [],

    "Precision": [],

    "Recall": [],

    "F1-Score": []

}


# ============================================================
# 9. START CROSS-VALIDATION
# ============================================================

print("\n" + "=" * 80)

print("STARTING PAIRED 5-FOLD EVALUATION")

print("=" * 80)


for fold, (

    train_indices,

    validation_indices

) in enumerate(

    sgkf.split(

        X_statistical,

        y_statistical,

        groups=groups_statistical

    ),

    start=1

):


    print("\n" + "=" * 80)

    print(
        f"FOLD {fold} / 5"
    )

    print("=" * 80)


    # ========================================================
    # FOLD DATA
    # ========================================================

    X_train_fold = X_statistical[
        train_indices
    ]


    X_validation_fold = X_statistical[
        validation_indices
    ]


    y_train_fold = y_statistical[
        train_indices
    ]


    y_validation_fold = y_statistical[
        validation_indices
    ]


    # ========================================================
    # CHECK GROUP OVERLAP
    # ========================================================

    training_groups = set(

        groups_statistical[
            train_indices
        ]

    )


    validation_groups = set(

        groups_statistical[
            validation_indices
        ]

    )


    overlap = (

        training_groups

        &

        validation_groups

    )


    print(
        "\nTraining Shape:",
        X_train_fold.shape
    )


    print(
        "Validation Shape:",
        X_validation_fold.shape
    )


    print(
        "Group Overlap:",
        len(overlap)
    )


    assert len(overlap) == 0


    # ========================================================
    # CALLBACK
    # ========================================================

    early_stopping = EarlyStopping(

        monitor="val_loss",

        patience=8,

        restore_best_weights=True,

        mode="min",

        verbose=0

    )


    # ========================================================
    # BASELINE MODEL
    # ========================================================

    print(
        "\nTraining Baseline BiLSTM..."
    )


    tf.keras.backend.clear_session()

    gc.collect()


    tf.random.set_seed(

        42 + fold

    )


    baseline_model = (

        create_baseline_bilstm()

    )


    baseline_model.fit(

        X_train_fold,

        y_train_fold,

        validation_data=(

            X_validation_fold,

            y_validation_fold

        ),

        epochs=50,

        batch_size=32,

        callbacks=[

            early_stopping

        ],

        verbose=0

    )


    # ========================================================
    # BASELINE PREDICTIONS
    # ========================================================

    baseline_probability = (

        baseline_model.predict(

            X_validation_fold,

            verbose=0

        ).ravel()

    )


    baseline_prediction = (

        baseline_probability >= 0.5

    ).astype("int32")


    # ========================================================
    # BASELINE METRICS
    # ========================================================

    baseline_accuracy = accuracy_score(

        y_validation_fold,

        baseline_prediction

    )


    baseline_precision = precision_score(

        y_validation_fold,

        baseline_prediction,

        zero_division=0

    )


    baseline_recall = recall_score(

        y_validation_fold,

        baseline_prediction,

        zero_division=0

    )


    baseline_f1 = f1_score(

        y_validation_fold,

        baseline_prediction,

        zero_division=0

    )


    baseline_scores["Accuracy"].append(

        baseline_accuracy * 100

    )


    baseline_scores["Precision"].append(

        baseline_precision * 100

    )


    baseline_scores["Recall"].append(

        baseline_recall * 100

    )


    baseline_scores["F1-Score"].append(

        baseline_f1 * 100

    )


    # ========================================================
    # DELETE BASELINE MODEL
    # ========================================================

    del baseline_model

    tf.keras.backend.clear_session()

    gc.collect()


    # ========================================================
    # FULL MODEL
    # ========================================================

    print(
        "Training TO-TBi-STMNet..."
    )


    tf.random.set_seed(

        42 + fold

    )


    full_model = (

        create_full_model()

    )


    full_model.fit(

        X_train_fold,

        y_train_fold,

        validation_data=(

            X_validation_fold,

            y_validation_fold

        ),

        epochs=50,

        batch_size=32,

        callbacks=[

            EarlyStopping(

                monitor="val_loss",

                patience=8,

                restore_best_weights=True,

                mode="min",

                verbose=0

            )

        ],

        verbose=0

    )


    # ========================================================
    # FULL MODEL PREDICTIONS
    # ========================================================

    full_probability = (

        full_model.predict(

            X_validation_fold,

            verbose=0

        ).ravel()

    )


    full_prediction = (

        full_probability >= 0.5

    ).astype("int32")


    # ========================================================
    # FULL MODEL METRICS
    # ========================================================

    full_accuracy = accuracy_score(

        y_validation_fold,

        full_prediction

    )


    full_precision = precision_score(

        y_validation_fold,

        full_prediction,

        zero_division=0

    )


    full_recall = recall_score(

        y_validation_fold,

        full_prediction,

        zero_division=0

    )


    full_f1 = f1_score(

        y_validation_fold,

        full_prediction,

        zero_division=0

    )


    full_model_scores["Accuracy"].append(

        full_accuracy * 100

    )


    full_model_scores["Precision"].append(

        full_precision * 100

    )


    full_model_scores["Recall"].append(

        full_recall * 100

    )


    full_model_scores["F1-Score"].append(

        full_f1 * 100

    )


    # ========================================================
    # DISPLAY FOLD RESULTS
    # ========================================================

    print("\nFold Results:")


    print(

        f"Baseline Accuracy : "

        f"{baseline_accuracy * 100:.4f}%"

    )


    print(

        f"Full Accuracy     : "

        f"{full_accuracy * 100:.4f}%"

    )


    # ========================================================
    # DELETE FULL MODEL
    # ========================================================

    del full_model

    tf.keras.backend.clear_session()

    gc.collect()


# ============================================================
# 10. FUNCTION TO CALCULATE 95% CONFIDENCE INTERVAL
# ============================================================

def calculate_confidence_interval(

    scores,

    confidence=0.95

):


    scores = np.asarray(

        scores,

        dtype=np.float64

    )


    n = len(scores)


    mean = np.mean(scores)


    standard_error = (

        np.std(

            scores,

            ddof=1

        )

        /

        np.sqrt(n)

    )


    critical_value = t.ppf(

        (1 + confidence) / 2,

        df=n - 1

    )


    margin_error = (

        critical_value

        *

        standard_error

    )


    lower_bound = (

        mean

        -

        margin_error

    )


    upper_bound = (

        mean

        +

        margin_error

    )


    return (

        mean,

        lower_bound,

        upper_bound

    )


# ============================================================
# 11. STATISTICAL ANALYSIS
# ============================================================

statistical_results = []


metrics = [

    "Accuracy",

    "Precision",

    "Recall",

    "F1-Score"

]


for metric in metrics:


    # --------------------------------------------------------
    # GET SCORES
    # --------------------------------------------------------

    baseline_values = np.asarray(

        baseline_scores[metric],

        dtype=np.float64

    )


    full_values = np.asarray(

        full_model_scores[metric],

        dtype=np.float64

    )


    # --------------------------------------------------------
    # BASELINE 95% CI
    # --------------------------------------------------------

    (

        baseline_mean,

        baseline_ci_lower,

        baseline_ci_upper

    ) = calculate_confidence_interval(

        baseline_values

    )


    # --------------------------------------------------------
    # FULL MODEL 95% CI
    # --------------------------------------------------------

    (

        full_mean,

        full_ci_lower,

        full_ci_upper

    ) = calculate_confidence_interval(

        full_values

    )


    # --------------------------------------------------------
    # PAIRED T-TEST
    # --------------------------------------------------------

    t_value, p_value = ttest_rel(

        full_values,

        baseline_values

    )


    # --------------------------------------------------------
    # HANDLE IDENTICAL RESULTS
    # --------------------------------------------------------

    if np.allclose(

        full_values,

        baseline_values

    ):


        t_value = 0.0

        p_value = 1.0


    # --------------------------------------------------------
    # SIGNIFICANCE DECISION
    # --------------------------------------------------------

    if p_value < 0.001:

        significance = "Highly Significant"


    elif p_value < 0.05:

        significance = "Significant"


    else:

        significance = "Not Significant"


    # --------------------------------------------------------
    # STORE RESULT
    # --------------------------------------------------------

    statistical_results.append({

        "Metric":

            metric.lower(),

        "Baseline (BiLSTM) Mean (%)":

            baseline_mean,

        "Baseline 95% CI":

            (
                f"[{baseline_ci_lower:.4f}, "
                f"{baseline_ci_upper:.4f}]"
            ),

        "TO-TBi-STMNet Mean (%)":

            full_mean,

        "TO-TBi-STMNet 95% CI":

            (
                f"[{full_ci_lower:.4f}, "
                f"{full_ci_upper:.4f}]"
            ),

        "t-value":

            t_value,

        "p-value":

            p_value,

        "Significance":

            significance

    })


# ============================================================
# 12. CREATE RESULTS DATAFRAME
# ============================================================

statistical_results_df = pd.DataFrame(

    statistical_results

)


# ============================================================
# 13. DISPLAY INDIVIDUAL FOLD SCORES
# ============================================================

print("\n\n" + "=" * 100)

print("INDIVIDUAL 5-FOLD SCORES")

print("=" * 100)


for metric in metrics:


    print(

        f"\n{metric}"

    )


    print(

        "Baseline:",

        np.round(

            baseline_scores[metric],

            4

        )

    )


    print(

        "TO-TBi-STMNet:",

        np.round(

            full_model_scores[metric],

            4

        )

    )


# ============================================================
# 14. DISPLAY FINAL STATISTICAL TABLE
# ============================================================

print("\n\n" + "=" * 160)

print("FINAL STATISTICAL SIGNIFICANCE ANALYSIS")

print("=" * 160)


print(

    statistical_results_df.to_string(

        index=False,

        formatters={

            "Baseline (BiLSTM) Mean (%)":

                lambda x: f"{x:.4f}",

            "TO-TBi-STMNet Mean (%)":

                lambda x: f"{x:.4f}",

            "t-value":

                lambda x: f"{x:.4f}",

            "p-value":

                lambda x: f"{x:.6f}"

        }

    )

)


# ============================================================
# 15. SAVE RESULTS
# ============================================================

statistical_results_df.to_csv(

    "bilstm_vs_to_tbistmnet_statistical_analysis.csv",

    index=False

)


# ============================================================
# 16. FINAL SUMMARY
# ============================================================

print("\n" + "=" * 80)

print("STATISTICAL ANALYSIS SUMMARY")

print("=" * 80)


for _, row in statistical_results_df.iterrows():


    print(

        f"\nMetric       : {row['Metric']}"

    )


    print(

        f"Baseline Mean: "

        f"{row['Baseline (BiLSTM) Mean (%)']:.4f}%"

    )


    print(

        f"Full Mean    : "

        f"{row['TO-TBi-STMNet Mean (%)']:.4f}%"

    )


    print(

        f"t-value      : "

        f"{row['t-value']:.4f}"

    )


    print(

        f"p-value      : "

        f"{row['p-value']:.6f}"

    )


    print(

        f"Significance : "

        f"{row['Significance']}"

    )


print("\nResults Saved As:")

print(

    "bilstm_vs_to_tbistmnet_statistical_analysis.csv"

)


print(

    "\nSTATISTICAL SIGNIFICANCE ANALYSIS COMPLETED SUCCESSFULLY"

)
```

```python
# ============================================================
# 5 INDEPENDENT RUNS PERFORMANCE EVALUATION
# TO-TBi-STMNet
# ============================================================

import numpy as np
import pandas as pd
import tensorflow as tf
import gc

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)


# ============================================================
# 1. CONFIGURATION
# ============================================================

NUMBER_OF_RUNS = 5

RANDOM_SEEDS = [42, 123, 456, 789, 1024]

EPOCHS = 50

BATCH_SIZE = 32


print("=" * 80)
print("5 INDEPENDENT RUNS PERFORMANCE EVALUATION")
print("=" * 80)

print("\nTraining Shape   :", X_sequence_train.shape)
print("Validation Shape :", X_sequence_validation.shape)
print("Testing Shape    :", X_sequence_test.shape)

print("\nTraining Target   :", y_train_ablation.shape)
print("Validation Target :", y_validation_ablation.shape)
print("Testing Target    :", y_test_ablation.shape)


# ============================================================
# 2. RESULT STORAGE
# ============================================================

run_results = []


# ============================================================
# 3. START 5 INDEPENDENT RUNS
# ============================================================

for run_number, random_seed in enumerate(

    RANDOM_SEEDS,

    start=1

):


    print("\n" + "=" * 80)

    print(
        f"RUN {run_number} / {NUMBER_OF_RUNS}"
    )

    print("=" * 80)


    # ========================================================
    # CLEAR PREVIOUS MODEL
    # ========================================================

    tf.keras.backend.clear_session()

    gc.collect()


    # ========================================================
    # SET DIFFERENT RANDOM SEED
    # ========================================================

    np.random.seed(random_seed)

    tf.random.set_seed(random_seed)


    print(
        "\nRandom Seed:",
        random_seed
    )


    # ========================================================
    # CREATE NEW FULL MODEL
    #
    # Uses the function created in the ablation code:
    #
    # create_transformer_bilstm()
    #
    # Uses best TO parameters:
    #
    # best_parameters
    # ========================================================

    run_model = create_transformer_bilstm(

        **best_parameters

    )


    # ========================================================
    # CALLBACKS
    # ========================================================

    early_stopping = EarlyStopping(

        monitor="val_loss",

        patience=8,

        restore_best_weights=True,

        mode="min",

        verbose=0

    )


    reduce_learning_rate = ReduceLROnPlateau(

        monitor="val_loss",

        factor=0.5,

        patience=4,

        min_lr=1e-6,

        mode="min",

        verbose=0

    )


    # ========================================================
    # TRAIN MODEL
    # ========================================================

    print(
        "\nTraining Model..."
    )


    history = run_model.fit(

        X_sequence_train,

        y_train_ablation,

        validation_data=(

            X_sequence_validation,

            y_validation_ablation

        ),

        epochs=EPOCHS,

        batch_size=BATCH_SIZE,

        callbacks=[

            early_stopping,

            reduce_learning_rate

        ],

        verbose=0

    )


    print(

        "Epochs Trained:",

        len(history.history["loss"])

    )


    # ========================================================
    # GENERATE TEST PREDICTIONS
    # ========================================================

    test_probabilities = run_model.predict(

        X_sequence_test,

        batch_size=BATCH_SIZE,

        verbose=0

    ).ravel()


    test_predictions = (

        test_probabilities >= 0.5

    ).astype("int32")


    # ========================================================
    # CALCULATE METRICS
    # ========================================================

    accuracy = accuracy_score(

        y_test_ablation,

        test_predictions

    ) * 100


    precision = precision_score(

        y_test_ablation,

        test_predictions,

        zero_division=0

    ) * 100


    recall = recall_score(

        y_test_ablation,

        test_predictions,

        zero_division=0

    ) * 100


    f1 = f1_score(

        y_test_ablation,

        test_predictions,

        zero_division=0

    ) * 100


    # ========================================================
    # STORE RESULTS
    # ========================================================

    run_results.append({

        "Run":

            f"Run {run_number}",

        "Accuracy (%)":

            accuracy,

        "Precision (%)":

            precision,

        "Recall (%)":

            recall,

        "F1-score (%)":

            f1

    })


    # ========================================================
    # DISPLAY RUN RESULTS
    # ========================================================

    print("\n" + "-" * 60)

    print(
        f"RUN {run_number} RESULTS"
    )

    print("-" * 60)


    print(
        f"Accuracy  : {accuracy:.4f}%"
    )


    print(
        f"Precision : {precision:.4f}%"
    )


    print(
        f"Recall    : {recall:.4f}%"
    )


    print(
        f"F1-score  : {f1:.4f}%"
    )


    # ========================================================
    # DELETE MODEL
    # ========================================================

    del run_model

    tf.keras.backend.clear_session()

    gc.collect()


# ============================================================
# 4. CREATE RESULTS DATAFRAME
# ============================================================

run_results_df = pd.DataFrame(

    run_results

)


# ============================================================
# 5. CALCULATE MEAN
# ============================================================

metric_columns = [

    "Accuracy (%)",

    "Precision (%)",

    "Recall (%)",

    "F1-score (%)"

]


mean_values = run_results_df[

    metric_columns

].mean()


# ============================================================
# 6. CALCULATE STANDARD DEVIATION
#
# ddof=1 gives sample standard deviation
# ============================================================

std_values = run_results_df[

    metric_columns

].std(

    ddof=1

)


# ============================================================
# 7. CREATE MEAN ± SD ROW
# ============================================================

mean_sd_row = {

    "Run": "Mean ± SD"

}


for metric in metric_columns:

    mean_sd_row[metric] = (

        f"{mean_values[metric]:.4f} "

        f"± "

        f"{std_values[metric]:.4f}"

    )


# ============================================================
# 8. CREATE FINAL TABLE
# ============================================================

final_run_table = run_results_df.copy()


# Format individual results

for metric in metric_columns:

    final_run_table[metric] = (

        final_run_table[metric]

        .map(

            lambda value:

            f"{value:.4f}"

        )

    )


# Add Mean ± SD

final_run_table = pd.concat(

    [

        final_run_table,

        pd.DataFrame(

            [mean_sd_row]

        )

    ],

    ignore_index=True

)


# ============================================================
# 9. DISPLAY FINAL RESULTS TABLE
# ============================================================

print("\n\n" + "=" * 100)

print(
    "5 INDEPENDENT RUNS RESULTS"
)

print("=" * 100)


print(

    final_run_table.to_string(

        index=False

    )

)


# ============================================================
# 10. DISPLAY NUMERICAL SUMMARY
# ============================================================

print("\n" + "=" * 80)

print(
    "MEAN AND STANDARD DEVIATION SUMMARY"
)

print("=" * 80)


for metric in metric_columns:


    print(

        f"{metric:<20}: "

        f"{mean_values[metric]:.4f} "

        f"± "

        f"{std_values[metric]:.4f}"

    )


# ============================================================
# 11. SAVE INDIVIDUAL NUMERIC RESULTS
# ============================================================

run_results_df.to_csv(

    "5_independent_runs_numeric_results.csv",

    index=False

)


# ============================================================
# 12. SAVE FINAL FORMATTED TABLE
# ============================================================

final_run_table.to_csv(

    "5_independent_runs_final_table.csv",

    index=False

)


# ============================================================
# 13. FINAL SUMMARY
# ============================================================

print("\n" + "=" * 80)

print(
    "FINAL 5-RUN PERFORMANCE SUMMARY"
)

print("=" * 80)


print(

    f"Accuracy  : "

    f"{mean_values['Accuracy (%)']:.4f} "

    f"± "

    f"{std_values['Accuracy (%)']:.4f}%"

)


print(

    f"Precision : "

    f"{mean_values['Precision (%)']:.4f} "

    f"± "

    f"{std_values['Precision (%)']:.4f}%"

)


print(

    f"Recall    : "

    f"{mean_values['Recall (%)']:.4f} "

    f"± "

    f"{std_values['Recall (%)']:.4f}%"

)


print(

    f"F1-score  : "

    f"{mean_values['F1-score (%)']:.4f} "

    f"± "

    f"{std_values['F1-score (%)']:.4f}%"

)


print("\nSaved Files:")

print(
    "1. 5_independent_runs_numeric_results.csv"
)

print(
    "2. 5_independent_runs_final_table.csv"
)


print(
    "\n5 INDEPENDENT RUNS EVALUATION COMPLETED SUCCESSFULLY"
)
```
